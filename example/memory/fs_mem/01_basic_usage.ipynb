# Basic Usage with FsMem
Create a task memory directory, update state, and search memory.

```python
import os, sys
sys.path.append(os.path.abspath('../../../src'))
from memory import FsMem
from memory.fs_mem import search_memory
from llm_provider import get_llm
mem_llm = get_llm(temperature=0.0)
```

## Prepare Memory Directory

```python
fs_mem_for_dir = FsMem(task_dir='./example_output/fs_mem_basic', llm=mem_llm)
state0 = {'messages':[{'role':'user','content':'Plan a small web app to track tasks.'}]}
task_dir = fs_mem_for_dir.create_task_dir(state0, base_path='./example_output/fs_mem_basic')
task_dir
```

## Update State

```python
fs_mem = FsMem(task_dir=task_dir, llm=mem_llm)
state = {'messages':[
    {'role':'user','content':'Plan a small web app to track tasks.'},
    {'role':'assistant','content':'Use FastAPI and SQLite.'}
]}
state = fs_mem.update_state(state, max_messages=2)
state = fs_mem.update_state(state, max_messages=2)
state['context'][:500]
```

## List Memory Files

```python
import os
knowledge_dir = os.path.join(task_dir, 'knowledge')
sorted(os.listdir(knowledge_dir)) if os.path.isdir(knowledge_dir) else []
```

## Search Memory

```python
search_memory('FastAPI', task_dir=task_dir)[:200]
```
