# Agent Integration with FsMem
Real LLM + BaseAgent and FMemAgent.

```python
import os, sys
sys.path.append(os.path.abspath('../../../src'))
from memory import FsMem
from memory.fs_mem import FMemAgent
from agent_blocks_hub.base_agent import create_base_agent
from llm_provider import get_llm
from langchain_core.messages import HumanMessage
task_llm = get_llm(temperature=0.2)
mem_llm = get_llm(temperature=0.0)
```

## Prepare Memory Directory

```python
fs_mem_for_dir = FsMem(task_dir='./example_output/fs_mem_agent', llm=mem_llm)
state0 = {'messages':[{'role':'user','content':'Build a REST API for tasks.'}]}
task_dir = fs_mem_for_dir.create_task_dir(state0, base_path='./example_output/fs_mem_agent')
task_dir
```

## BaseAgent with FsMem

```python
fs_mem = FsMem(task_dir=task_dir, llm=mem_llm)
agent = create_base_agent(llm=task_llm, tools=None, system_prompt='You are a helpful assistant.')
resp = agent.invoke({'messages':[{'role':'user','content':'Choose a framework.'}]}, memory=fs_mem)
getattr(resp, 'content', resp)
```

## FMemAgent call_with_memory

```python
fmem_agent = FMemAgent(task_dir=task_dir)
resp2 = fmem_agent.call_with_memory({'messages':[HumanMessage(content='Pick a DB for the app')]}, llm=task_llm, system_prompt='You are a helpful assistant.')
getattr(resp2, 'content', resp2)
```
